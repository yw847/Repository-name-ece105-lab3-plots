```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

rng = np.random.default_rng(seed=1234)  # Replace 1234 with last 4 digits of your Drexel ID

# Generate sensor readings
sensor_a = rng.normal(loc=25, scale=3, size=200)
sensor_b = rng.normal(loc=27, scale=4.5, size=200)

# Generate timestamps uniformly from 0 to 10 seconds
timestamps = np.linspace(0, 10, 200)
```

```python
# Create scatter plot of sensor readings\n
plt.figure(figsize=(10, 6))\n
plt.scatter(timestamps, sensor_a, alpha=0.6, label='Sensor A', s=30)\n
plt.scatter(timestamps, sensor_b, alpha=0.6, label='Sensor B', s=30)\n
plt.xlabel('Time (seconds)')\n
plt.ylabel('Temperature (°C)')\n
plt.title('Temperature Readings Over Time')\n
plt.legend()\n
plt.grid(True, alpha=0.3)\n
plt.show()
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(timestamps, sensor_a, color='blue', alpha=0.6, label='Sensor A', s=30)
plt.scatter(timestamps, sensor_b, color='orange', alpha=0.6, label='Sensor B', s=30)
plt.xlabel('Time (seconds)')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Readings Over Time')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.
```

```python
plt.figure(figsize=(10, 6))
plt.hist(sensor_a, bins=30, alpha=0.5, label='Sensor A', color='blue')
plt.hist(sensor_b, bins=30, alpha=0.5, label='Sensor B', color='orange')

# Add vertical lines at the means
plt.axvline(sensor_a.mean(), color='blue', linestyle='--', linewidth=2, label=f'Mean A: {sensor_a.mean():.2f}°C')
plt.axvline(sensor_b.mean(), color='orange', linestyle='--', linewidth=2, label=f'Mean B: {sensor_b.mean():.2f}°C')

plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Temperature Distribution - Sensor A vs Sensor B')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.

```python
plt.figure(figsize=(10, 6))
plt.boxplot([sensor_a, sensor_b], labels=['Sensor A', 'Sensor B'])

# Calculate overall mean of both sensors
overall_mean = np.concatenate([sensor_a, sensor_b]).mean()
plt.axhline(overall_mean, color='red', linestyle='--', linewidth=2, label=f'Overall Mean: {overall_mean:.2f}°C')

plt.ylabel('Temperature (°C)')
plt.title('Temperature Distribution Comparison')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
